```python
import numpy as np
import matplotlib.pyplot as plt
#from skimage.io import imshow, imread
#from skimage.color import rgb2hsv, hsv2rgb
import cv2
```

```python
#Importa arquivo a ser tratado
from google.colab import files
imagem = files.upload()
filename = next(iter(imagem))
```

```python
face = cv2.imread(filename)
img = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Convert BGR to HSV and parse HSV
img_hsv = cv2.cvtColor(face, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```

```python
#Hue Component
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(img_hsv[:,:,0], cmap='hsv')
plt.colorbar();
```

```python
#Saturation
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(img_hsv[:,:,1], cmap='gray')
plt.colorbar();
```

```python
#Value
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(img_hsv[:,:,2], cmap='gray')
plt.colorbar();
```
